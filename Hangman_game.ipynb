```python
import random

def hangman():
    # List of game words, their hints, and associated riddles
    word_list = [
        ("python", ["A popular programming language.", "Named after a type of snake.", "Often used for AI and web development."],
         "I’m a programming language, but also a slithering reptile. What am I?"),
        ("programming", ["What developers do.", "Involves writing code.", "Can be a fun or challenging career."],
         "I’m the process of telling a computer what to do. What am I?"),
        ("hangman", ["The game you are playing now!", "Involves guessing letters.", "A word-guessing game with a gallows."],
         "I’m a word-guessing game that might leave you hanging. What am I?"),
        ("listen", ["What you should do carefully.", "Opposite of talking.", "Involves using your ears."],
         "I’m something you do with your ears to understand someone. What am I?"),
        ("class", ["I’m a room where students learn. What am I?", "Students spend most of their day here.", "A teacher leads the lessons in this place."],
         "I'm full of learning but never take a test; what am I?"),
        ("follow", ["To come after something.", "You might do this on social media.", "Opposite of leading."],
         "I’m something you do to stay updated with someone online. What am I?")
    ]

    while True:
        # Randomly select a word, hints, and its riddle
        word, hints, riddle = random.choice(word_list)

        # Initialize game variables
        guessed_word = ["_" for _ in word]  # Display word as underscores
        guessed_letters = {}  # Dictionary to track guessed letters and their counts
        incorrect_letters = set()
        incorrect_guesses = 0
        max_incorrect_guesses = 5
        hints_used = 0
        max_hints = 3

        print("Welcome to Hangman!")
        print(f"\nRiddle: {riddle}")
        print("You have", max_incorrect_guesses, "chances to guess the word.")

        while incorrect_guesses < max_incorrect_guesses:
            print("\nWord:", " ".join(guessed_word))
            print("Guessed letters:", ", ".join(sorted(guessed_letters.keys())))
            print("Incorrect guesses left:", max_incorrect_guesses - incorrect_guesses)

            # User input
            guess = input("Guess a letter or type 'hint' for a clue: ").lower()

            if guess == "hint":
                if hints_used < max_hints:
                    print(f"Hint {hints_used + 1}: {hints[hints_used]}")
                    hints_used += 1
                else:
                    print("\033[91mYou have reached the maximum number of hints!\033[0m")
                continue

            if len(guess) != 1 or not guess.isalpha():
                print("\033[91mPlease enter a single valid letter.\033[0m")
                continue

            # Check if the letter has already been guessed
            if guess in guessed_letters:
                guessed_letters[guess] += 1  # Increment the count for the guessed letter

                # Warn initially before reducing chances
                if guessed_letters[guess] <= 2:
                    print(f"\033[91mYou already guessed that letter. ({guessed_letters[guess]} times)\033[0m")
                else:
                    print(f"\033[91mYou already guessed that letter too many times. Reducing a chance.\033[0m")
                    incorrect_guesses += 1
                continue
            else:
                guessed_letters[guess] = 1  # Add the guessed letter to the dictionary 

            if guess in word:
                print("\033[91mGood job! That letter is in the word.\033[0m")
                for i, letter in enumerate(word):
                    if letter == guess:
                        guessed_word[i] = guess

                if "_" not in guessed_word:
                    print("\033[91m\nCongratulations! You guessed the word:\033[0m", word)
                    break
            else:
                print("\033[91mOops! That letter is not in the word.\033[0m")
                incorrect_letters.add(guess)
                incorrect_guesses += 1

        if incorrect_guesses == max_incorrect_guesses:
            print("\nGame Over! The word was:", word)

        # Ask if the user wants to play again
        while True:
            play_again = input("\nDo you want to play again? (yes or no): ").lower()
            if play_again == "yes":
                break
            elif play_again == "no":
                print("Thanks for playing Hangman. Goodbye!")
                return
            else:
                print("Please enter 'yes' or 'no'.")



```

```python
hangman()
```
